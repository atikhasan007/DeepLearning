```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Input → Linear → ReLU → Linear → Sigmoid → Output (0–1)
# option 1 (vreate nn modules)
class NeuralNet(nn.Module):
  def __init__(self , input_size , hidden_size):
    super(NeuralNet , self).__init__() # “parent class (nn.Module) এর constructor আগে run করো”
    self.linear1 = nn.Linear(input_size , hidden_size)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(hidden_size, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self , x):
    out = self.linear1(x)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
# option 2 (use activation functions directly in forward pass)
# input -> Linear1 -> ReLU -> Linear2 -> Sigmoid -> Output (0-1)
class NeuralNet(nn.Module):
  def __init__(self, input_size , hidden_size):
    super(NeuralNet, self).__init__() # model initialize করা
    self.linear1 = nn.Linear(input_size, hidden_size)  # input → hidden layer
    self.linear2 = nn.Linear(hidden_size , 1)

  def forward(self, x): # data কীভাবে যাবে সেটা define
    out = torch.relu(self.linear1(x)) # Layer + activation একসাথে   linear1 → ReLU
    out = torch.sigmoid(self.linear2(out))  # linear2 → Sigmoid
    return out


```
