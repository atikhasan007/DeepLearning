*PyTorch এ autograd এর কাজ হলো automatic differentiation করা, অর্থাৎ automatically gradient calculate করা।*





1. Operations track করে

যে tensor এ requires_grad=True থাকে, তার সব calculation graph আকারে save হয়।

2. Gradient calculate করে

backward() call করলে chain rule ব্যবহার করে gradient বের করে।

3. Neural Network train করতে সাহায্য করে

Training এর সময়:

* Forward pass
* Loss calculate
* Backpropagation
* Weight update

এই পুরো process এ autograd gradient দেয়।

```python
# rand range [0 -> 1] , Uniform  distribution
# randn range [-enfinity , +enfinity] mean = 0 , std = 1
```

* rand() → random initialization, probability, dropout mask
* randn() → neural network weight initialization (সবচেয়ে common)

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)
```

```python
y = x + 2
```

```python
print(y)
```

```python
z = y*y*2
print(z)
```

```python
z = z.mean()
print(z)
```

```python
z.backward() # dz / dx
print(x.grad)
```

```python
 # x.requires_grad(False)   “এই tensor-এর জন্য gradient হিসাব করো না
 # x.detach() এটা tensor-কে পুরো computation graph থেকে আলাদা করে দেয়
 # with torch.no_grad():এটা একটা context manager, যার ভিতরে সব gradient tracking বন্ধ
```

```python
x.requires_grad_(True)
```

```python
x.requires_grad_(False)
print(x)
```

```python
y = x.detach()
print(y)
```

```python
y = x + 2
print(y)
```

```python
with torch.no_grad():
  y = x + 2
  print(y)
```

# **Training Example**

```python
import torch

weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
  model_output = (weights * 3).sum()

  model_output.backward()

  print(weights.grad)
```

```python
# empty gradient

import torch

weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
  model_output = (weights * 3).sum()

  model_output.backward()

  print(weights.grad)
  weights.grad.zero_()
```

# **Optimization**

```python
import torch

weights = torch.ones(4, requires_grad=True)
optimizer = torch.optim.SGD([weights], lr=0.01) # weights should be in a list
optimizer.step()
optimizer.zero_grad()
```
