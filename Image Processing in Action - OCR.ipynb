```
#draw the bounding boxes
import easyocr

#this needs to run only once to load the model into memory
reader = easyocr.Reader(['en'])
bounding_box = reader.readtext("/content/bb1e32e98b2a7fff20a38f3e2b350871.jpg")
bounding_box
```

```
#draw bounding boxes

import PIL
from PIL import ImageDraw
def draw_boxes(image, bounding_points, color='red',width=2):
  draw = ImageDraw.Draw(image)
  for bound in bounding_points:
    p0, p1, p2, p3 = bound[0] # Unpack the four points from the bounding box
    draw.line([*p0, *p1, *p2,*p3,*p0], fill=color, width=width) # Use fill instead of file for color
  return image
```

```
draw_boxes(img,bounding_box)
```

```
import PIL
from PIL import ImageDraw
import easyocr
import cv2


#this needs to run only one load the model into memory
bn_reader = easyocr.Reader(['bn'])


def draw_boxes(image, bounding_points, color='green', width=2):
  draw = ImageDraw.Draw(image)
  for bound in bounding_points:
    p0, p1, p2, p3 = bound[0]
    draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
  return image

```

```
#read the photo

bn_img = PIL.Image.open("/content/images.jpeg")
bounding_box = bn_reader.readtext(bn_img)

#draw the bounding boxes
draw_boxes(bn_img, bounding_box)
```

```
# Extracting only the text from the Bangla results
bangla_sentences = [item[1] for item in bangla_result]

# Displaying the extracted Bangla sentences
for sentence in bangla_sentences:
    print(sentence)
```
