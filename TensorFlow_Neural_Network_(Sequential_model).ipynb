```python
import tensorflow as tf
from tensorflow import  keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

```

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
fashion_mnist # shirt , T-shirt ect data
```

```python
fashion_mnist.load_data()
```

```python
(train_image, train_labels), (test_image, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
print(class_names)
```

```python
#preprocess
```

```python
train_image = train_image / 255.0
test_image = test_image / 255.0
```

```python
train_image[0,23,23]
```

```python
temp_var = 194/255
temp_var
```

```python
# model
```

```python
model = keras.Sequential (

     [
         keras.layers.Flatten(input_shape=(28,28)),
         keras.layers.Dense(128, activation='relu'),
         keras.layers.Dense(10, activation='softmax')

     ]

)
```

```python
model.compile(
    optimizer='adam',          # শেখার পদ্ধতি
    loss='sparse_categorical_crossentropy',  # ভুল কমানোর মান
    metrics=['accuracy']       # পারফরম্যান্স দেখা
)
```

```python
model.fit(train_image, train_labels, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(test_image, test_labels, verbose=1)
print("Accuracy : ", test_acc)
```

```python
print("test Loss : ", test_loss)
```
